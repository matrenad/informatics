```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n-1):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

def k_smallest():
    n, k = map(int, input().split())
    num = list(map(int, input().split()))
    print(" ".join(map(str, bubble_sort(num)[:k])))

k_smallest()
```

```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    while len(left) > 0 and len(right) > 0:
        if left[0] <= right[0]:
            result.append(left.pop(0))
        else:
            result.append(right.pop(0))
    result.extend(left)
    result.extend(right)
    return result

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >= 0 and arr[j] > key:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key
    return arr

def selection_sort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def k_smallest():
    n, k = map(int, input().split())
    num = list(map(int, input().split()))
    
    print("Bubble Sort:")
    print(" ".join(map(str, bubble_sort(num)[:k])))

    print("Merge Sort:")
    print(" ".join(map(str, merge_sort(num)[:k])))

    print("Insertion Sort:")
    print(" ".join(map(str, insertion_sort(num)[:k])))

    print("Selection Sort:")
    print(" ".join(map(str, selection_sort(num)[:k])))

k_smallest()
```

# Задача 2

7 - размер очереди, то есть количество клиентов 3 - количество касс
1 2 3 4 5 3 1 - время на обслуживание каждого из клиентов

Как только один из клиентов завершает процесс, следующий в очереди подходит к освободившейся кассе. 

```python
def queue():
    n, m = map(int, input().split())
    customers = list(map(int, input().split()))
    tills = [0] * m
    served_customers = [[] for i in range(m)]
    z=1
    for customer in customers:
        print(f"Шаг номер {z}")
        z+=1
        next_till = tills.index(min(tills))
        print(f"Сумма времени чтобы обслужить подходивших клиентов {tills}")
        print(f" Следующая касса, к которой подойдут {next_till}")
        tills[next_till] += customer
        print(f"Готовый подойти клиент {customer}")
        served_customers[next_till].append(customer)
        print(f"Обслуженные кассой клиенты {served_customers}")
    total_time = [sum(customers) for customers in served_customers]
    return max(total_time)

queue()
```

# Задача 3

Бинарное дерево поиска — это дерево, для которого выполняются следующие свойства:

У каждой вершины не более двух детей.

Все вершины обладают ключами, на которых определена операция сравнения ( то есть тип данных числа или строки).

У всех вершин левого поддерева вершины v ключи не больше, чем ключ v.

У всех вершин правого поддерева вершины v ключи больше, чем ключ v.

Оба поддерева — левое и правое — являются двоичными деревьями поиска.

Основное преимущество бинарных деревьев поиска в том, что в них можно легко производить поиск элементов.

Олимпиадная задачка)

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def insert(root, value):
    if root is None:
        return Node(value)
    if value < root.value:
        print(f"Добавляем {value} в левое поддерево у родителя {root.value}")
        root.left = insert(root.left, value)
        print(f"Левый потомок {root.left.value if root.left else None} у родителя {root.value}")
    elif value > root.value:
        print(f"Добавляем {value} в правое поддерево у родителя {root.value}")
        root.right = insert(root.right, value)
        print(f"Правый потомок {root.right.value if root.right else None}  у родителя {root.value} ")
    return root

def in_order_traversal(root):
    if root:
        in_order_traversal(root.left)
        if root.left is None and root.right is None:
            print(root.value, end=" ")
        in_order_traversal(root.right)

n = int(input())
nums = list(map(int, input().split()))

root = None
for num in nums:
    root = insert(root, num)

in_order_traversal(root)
```
